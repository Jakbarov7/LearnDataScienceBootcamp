```python
import pandas as pd
```

```python
# DF lar yaratib olamiz
data = {
    'Brand': ['Mercedes', 'BMW', 'Mercedes', 'BMW', 'Mercedes', 'BMW'],
    'Model': ['A-Class', '3 Series', 'C-Class', '5 Series', 'E-Class', '7 Series'],
    'Year': [2019, 2018, 2020, 2017, 2021, 2016],
    'Price': [35000, 32000, 40000, 45000, 50000, 60000]
}
df = pd.DataFrame(data)
df
df1 = df[df['Brand'] == 'Mercedes']
df2 = df[df['Brand'] == 'BMW']
```

```python
df1
```

```python
df2
```

```python
# yangi HDF fayl yaratamiz
hdf_dfs = pd.HDFStore("brandCars.h5")
```

```python
# faylga dataFramelarni yuklimiz
hdf_dfs.put("mercadez", df1)
hdf_dfs.put("bmw", df2)
```

```python
# keylarni ko'ramiz
hdf_dfs.keys()
```
