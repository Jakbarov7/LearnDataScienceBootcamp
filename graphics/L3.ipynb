```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# DF yuklimiz
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns = ['maydon','aholi']
df
```

```python
# barplot da grafik chizish
sns.barplot(x=df.index, y=df['aholi'])
plt.show()
```

```python
# plt funksiyalaridan foydalanib grafikni chiroyli qilamiz
plt.figure(figsize=(12, 6))
plt.grid()
plt.title("Viloyatlarning aholisi")
plt.xlabel("Viloyat nomlari")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90)
sns.barplot(x=df.index, y=df['aholi'])
plt.show()
```
