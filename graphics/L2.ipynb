```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# DF o'qiymiz
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
# seaborn bilan grafik chizamiz
sns.lineplot(data=df)
plt.show()
```

```python
# shu grafikni matplotlibda chizb ko'ramiz
plt.plot(df)
plt.show()
```

```python
# seaborn bilan chizilgan grafikni yanada tushunarliroq qilamiz
plt.figure(figsize=(11, 5))
plt.title('Muzeylar grafigi')
plt.grid()
plt.xlabel("Date")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90) # x-o'qidagi sonlarni tartibliydi
sns.lineplot(data=df)
plt.show()
```

```python
# faqat bitta muzeyni grafigini chizish
plt.figure(figsize=(10, 5))
plt.title('Muzey grafigi')
plt.grid()
plt.xlabel("Date")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90) # x-o'qidagi sonlarni tartibliydi
sns.lineplot(data=df['Avila Adobe'])
plt.show()
```

```python
# bir vaqtda 2 va undan ko'p muzeylarni grafigini chizish
plt.figure(figsize=(10, 5))
plt.title('Muzey grafigi')
plt.grid()
plt.xlabel("Date")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90)
sns.lineplot(data=df[['Avila Adobe', 'Firehouse Museum']])
plt.show()
```

```python
# Chinese American Museum	va America Tropical Interpretive Center muzeylarini grafigini ko'ramiz
plt.figure(figsize=(12, 6))
plt.title('Muzey grafigi')
plt.grid()
plt.xlabel("Date")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90)
sns.lineplot(data=df[['Chinese American Museum', 'America Tropical Interpretive Center']])
plt.show()
```
