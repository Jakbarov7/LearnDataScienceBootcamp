```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**matplotlib kutubxonasi bilan ishlash**

```python
# grafik chizish
x = np.arange(1, 11)
plt.plot(x)
plt.show()
```

```python
# plotga ikki qiymat berish bilan grafik chizish
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# plot argumentlari bilan ishlash
plt.plot(x, y, linewidth=3) # chiziq qalinligi
plt.show()
```

```python
# grafik dizayni
plt.plot(x, y, linewidth=3, linestyle='--')
plt.show()
```

```python
# grafik chizig'ini rangi
plt.plot(x, y, linewidth=3, linestyle='-.', color='r')
plt.show()
```

```python
plt.plot(x, y, linewidth=3, linestyle='-.', color='b')
plt.show()
```

```python
# rangini shaffofligi
plt.plot(x, y, linewidth=2.5, linestyle=':', color='g', alpha=0.5)
plt.show()
```

```python
# marker argumenti
plt.plot(x, y, linewidth=3, linestyle='--', color='b', alpha=0.8, marker='^')
plt.show()
```

```python
# qo'shimcha funksiyalar | grafikka nom berish
plt.title("sin(x) grafigi")
plt.plot(x, y, linewidth=2.5, linestyle='--', alpha=0.7, marker='.', color='b')
plt.show()
```

```python
# x-o'qiga va y-o'qiga nomlar berish
plt.title("sin(x) grafigi")
plt.plot(x, y, linewidth=2.5, linestyle='--', alpha=0.7, marker='.', color='b')
plt.xlabel("x-o'qi")
plt.ylabel("y-o'qi")
plt.show()
```

```python
# grafik ustiga setka tortish va uning argumentlari
plt.title("sin(x) grafigi")
plt.grid()
plt.plot(x, y, linewidth=2.5, linestyle='--', alpha=0.7, marker='.', color='b')
plt.show()
```

```python
plt.title("sin(x) grafigi")
plt.grid(color='r', linestyle='--')
plt.plot(x, y, linewidth=2.5, linestyle='--', alpha=0.7, marker='.', color='b')
plt.show()
```

```python
# grafik rasmini katta yoki kichik qilish
plt.figure(figsize=(10, 4))
plt.grid(color='r', linestyle='--')
plt.plot(x, y, linewidth=2.5, linestyle='--', alpha=0.7, marker='.', color='b')
plt.show()
```

```python
# rasmni saqlash
plt.savefig("sin(x)")
```

```python
# sin(x) va cos(x) grafiklarini ustma-ust qilib chizamiz
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

# grafik hajmini beramiz
plt.figure(figsize=(12, 6))

# sin(x) grafigini chizamiz
plt.plot(x, sin, linewidth=2.5, linestyle='--', color='#8E44AD', alpha=0.7, marker='.')

# cos(x) grafigini chizamiz
plt.plot(x, cos, linewidth=2.5, linestyle='--', color='c', alpha=0.6, marker='.')

# grafikka umumiy nom beramiz
plt.title("sin va cos grafiklari")

# yangi funksiya | grafik rasmini bir burchagida qaysi rangli chiziqlar nimaniki ekanligini ko'rsatib turadi
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')

# grafik ustiga setka tortamiz
plt.grid(color='tomato', linestyle='--')
plt.show()
```
