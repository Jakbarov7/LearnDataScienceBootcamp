```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv")
```

```python
df.head()
```

```python
# mavjud bo'lmagan NaN qiymatlarni aniqlash usullari
df.isnull() # NaN bo'lgan qiymatlarni True bilan ko'rsatadi
```

```python
df.notnull() # NaN bo'lganlarni False qilib ko'rsatadi
```

```python
# jadval ustunlarining har biridan NaN qiymatlar nechtaligini aniqlash
df.isnull().sum()
```

```python
# jadvalimizni necha foizida qiymatlar yo'qligini aniqlash
df.isnull().sum()/len(df)*100
```

```python
# NaN lardan tashqari qancha qiymatlar borligini aniqlash
len(df.dropna())
```

```python
df.shape
```

```python
# NaN qiymatlari bo'lgan biror ustunlarni tashavorish
df.dropna(axis=1)
```

```python
# NaN qiymatlarni o'chirishda chegara qo'yishimiz mumkin
# Masalan : 50 tadan ko'p NaN ga ega bo'lgan qator yoki ustunlarni o'chirish , bunda 50 tadan kam NaN bo'lganlari o'chmiydi
df.dropna(axis=1, thresh=50).sample(20)
```

```python
# NaN qiymatlari bo'lgan aynan qaysi ustunlarni qatorlarini tashavorishni belgilash
df.dropna(subset=['price'])
```
