```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
```

```python
df.head()
```

```python
# barcha NaN qiymatlarni 0 yoki boshqa son bilan to'ldirish
df.fillna(0)
```

```python
df.fillna(-1)
```

```python
# aynan qaysi ustunlarni NaN qiymatlarini qanday qiymat bilan to'ldirish kerakligini berish
df.fillna({'horsepower':1, 'engine-type':0})
```

```python
# bir ustunni o'rta qiymatini o'sha ustunni NaN qiymatlari o'rniga yozish
df.fillna({'horsepower':np.mean(df['horsepower'])})
```

```python
df
```

```python
# yuqoridagi amalni birdaniga ikki yoki undan ko'p ustunlar uchun qilish
df.fillna({'horsepower':np.mean(df['horsepower']), 'length':np.mean(df['length'])})
```

```python
# NaN qiymatlarni o'zidan orqadagi qatorni qiymati bilan to'ldirish
df.fillna(method='ffill')
```

```python
df
```

```python
df.ffill()
```

```python
# NaN qiymatni o'zidan bitta keyingi qatorni qiymati bilan to'ldirvorish
df.bfill()
```
