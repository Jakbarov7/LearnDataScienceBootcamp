```python
import numpy as np
import pandas as pd
```

```python
# DF yaratamiz
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

df2 = pd.DataFrame({
    'D':[10, 11, 12],
    'B':[13, 14, 15],
    'C':[16, 17, 18]
})
```

```python
df1
```

```python
df2
```

```python
# DataFrame larni qo'shish
df1.add(df2) # buni o'rniga df1+df2 qisegam bo'lardi lekin metod ishlatishni foydali tarafini pasda ko'ramiz
```

```python
# NaN larni to'ldirish
df1.add(df2, fill_value=0) # ustunlarni o'zlarini qiymatlariga 0 qo'shib yozadi
```

Ayirishda ehtiyot bo'lish kerak . Ya'ni df1 bilan df2 ni joylarini almashtirvormasligimiz kerak , bu katta xatoga olib keladi .

```python
# DF larni bir-biridan ayirish
df2.sub(df1)
```

```python
# NaN larni to'ldirvoramiz
df2.sub(df1, fill_value=0)
```

```python
# df1 dan df2 ni ayirishni ko'ramiz
df2.rsub(df1, fill_value=0) # yoki df1.sub(df2) qilamiz
```

```python
# DF larni bo'lish | div
df2.div(df1, fill_value=0)
```

```python
# rdiv bundayam bor
df2.rdiv(df1, fill_value=0)
```

```python
# Series va DF o'rtasidagi arifmetik amallarda ham yuqoridagi metodlarni qo'llasak bo'ladi
df = pd.DataFrame(np.arange(1,13.).reshape(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
obj = pd.Series([1,1,2,2], index=df.columns)
```
