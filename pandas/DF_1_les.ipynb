```python
import numpy as np
import pandas as pd
```

```python
# Python_dictionary yordamida DataFrame(DF) yaratish
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       "Temp":[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]}
df = pd.DataFrame(data)
df
```

```python
# jadvalning yuqorisidan N ta qator ko'rish
df.head()
```

```python
df.head(3)
```

```python
# jadvalning hajmini aniqlash (qator va ustunlar sonini)
df.shape
```

```python
# jadal elementlari sonini aniqlash
df.size
```

```python
# Py_dct yordamida yaratilgan DF ning ustunlar tartibini o'zgartirish
df2 = pd.DataFrame(data, columns=['Temp', 'Yil', 'Aholi soni', 'YAIM']) # yangi ustun qo'shvoramiz
df2
```

```python
# DF ning indekslariga yangi qiymatlar berish
df2 = pd.DataFrame(data, columns=['Temp', 'Yil', 'YAIM', 'Aholi soni'], index=[21, 20, 19, 18, 17, 16, 15, 10])
df2
```

```python
# DF ning bir ustunini Series qilib olishning ikki usuli : 1-usul
df.Temp
```

```python
# 2-usul
df['Aholi soni']
```

```python
# yuqoridagi usul bilan bir ustunni barcha qiymatlarini bitta elementga o'zgartirish
df2.YAIM = 0
df2
```

```python
# Series yordamida DF ning bir qancha qatorlarini qiymatlarini o'zgartirish , bunda kerakli indekslar beriladi
sr = pd.Series([12.3, 11.2, 91.5, 18.6], index=[20, 18, 16, 10])
df2.YAIM = sr
df2
```

```python
df2.shape
```

```python
# List yordamida DF ning bir ustunini barcha qiymatlarini yangi elementlarga o'zgartirish
lst = [10.8, 11.2, 12.4, 13.5, 14.9, 15.3, 16.1, 17.8]
df2.YAIM = lst
df2
```

```python
# DF ga yangi ustun qo'shish
df2['new_ustun'] = 0
df2
```

```python
# Py_dct kabi biror ustunni o'chirish mumkin
del df2['new_ustun']
df2
```

```python
df2
```

```python
# DataFrame da filtrlash | 2018dan yuqori ma'lumotlarni olamiz
df2[df2.Yil > 2018]
```

```python
# yil 2018 dan kichik aholi soni 30 milliondan oshiqlarini olamiz
df2[(df2.Yil < 2018) & (df2['Aholi soni'] > 30)]
```

```python
# Nested dict yordamida DF yaratish
data2 = {
    'Uzbekistan':{2021:33.9, 2020:33.5, 2019:32.9, 2018:32.5, 2017:31.9},
    'Kazakhstan':{2021:18.9, 2020:18.7, 2019:18.5, 2018:18.3, 2017:18.1, 2016:17.8}
}
df3 = pd.DataFrame(data2)
df3
```

```python
# Series kabi DF indekslariga nom berish
df3.index.name = 'Yillar'
df3
```

```python
# Jadvaldan faqatgina bir necha qatorlarni qiymatlarini olish , bunda kerakli indekslarni beramiz
df4 = pd.DataFrame(data2, index=[2020, 2018, 2016])
df4
```

```python
# Transpose bilan indeks va ustunlar o'rnini almashtirish
df4.T
```
