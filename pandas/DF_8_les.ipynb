```python
import numpy as np
import pandas as pd
```

- `.loc[]` - indeks va ustun nomini berish orqali kerakli qiymatlarni olish
- `.iloc[]`- kerakli qiymatlarni olish uchun indeks va ustun tartib raqamlaridan foydalanish
- `.at[]` - faqat bitta qiymat olish , indeks va ustun nomi beriladi
- `.iat[]` - `at[]` bilan bir xil faqat bunda indeks va ustun tartib raqamlari beriladi

```python
data = {
    'USA': {
        2020: 331002651,
        2021: 331893745,
        2022: 332915073,
        2023: 334035655,
        2024: 335154543
    },
    'India': {
        2020: 1380004385,
        2021: 1393409038,
        2022: 1407563842,
        2023: 1421339190,
        2024: 1433783686
    },
    'China': {
        2020: 1444216107,
        2021: 1448471400,
        2022: 1453780764,
        2023: 1459453821,
        2024: 1465837300
    },
    'Brazil': {
        2020: 212559417,
        2021: 213993437,
        2022: 215353593,
        2023: 216688182,
        2024: 217999573
    },
    'Nigeria': {
        2020: 206139589,
        2021: 208679114,
        2022: 211400708,
        2023: 214028302,
        2024: 216746934
    }
}
df = pd.DataFrame(data)
df
```

```python
# DataFrame(DF) ning biror qatorini to'liq kesib olish
df.loc[[2022]] # 2022-indeksidagi ma'lumotlarni olamiz
```

```python
# ikki va undan ortiq indekslarni kesib olish
df.loc[[2024, 2023]]
```

```python
# kerakli ustun va indekslarni olish | 1-indekslar 2-ustun_nomlari yoziladi
df.loc[[2020, 2022], ['USA', 'Brazil']]
```

```python
df
```

```python
# Yuqoridagi misollarni indeks tartib raqamlari bo'yicha olish
df.iloc[[1]] # 2021-indeksni olamiz
```

```python
# kerakli ustun va qatorlarni olamiz
df.iloc[[0, 2], [1, 4]] # 2020-2022, India, Nigeryani olamiz
```

```python
# DF ning faqat birdona qiymatini olish | indeks va ustun nomi beriladi
df.at[2022, 'India'] # oddiy son qilib oberadi
```

```python
type(df.at[2022, 'India'])
```

```python
# yuqoridagi misolni indeks_tartib_raqamlari bo'yicha olish
df.iat[3, 1] # 2023-India qiymatini oberadi
```

```python
type(df.iat[3, 1])
```
