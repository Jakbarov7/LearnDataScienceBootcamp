```python
import pandas as pd
```

Reytinglash(ranking)

```python
# DF qabul qilamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
# DF ning biror ustuni bo'yicha reytinglash . Mashinani yili bo'yicha reytinglab natijalarni yangi Rank ustuniga berib ketamiz
df['ranking'] = df['year'].rank() # yil ustuni bo'yicha tartiblimiz
df.sort_values(by='year')
# mashinalardan bir xil yildagilar qancha ko'p bo'lsa reytingi shuncha kamayib ketadi
```

```python
# yuqoridagi misolni minimumdan maksimumga qarab reytinglimiz
df['year'].rank(method='min')
df.sort_values(by='year')
```

```python
# narx ustunini o'rtacha qiymati bo'yicha reytinglimiz
df['price'].rank(method='average')
df.sort_values(by='price')
```
