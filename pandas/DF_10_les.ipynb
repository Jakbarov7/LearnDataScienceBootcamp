```python
import numpy as np
import pandas as pd
```

DataFrame(DF) ga funksiyalarni qo'llash

```python
# DF yaratamiz
dct = {
    'Yillar':[2024, 2023, 2022, 2021, 2020, 2019, 2018],
    'Narxlar':[2.3, 4.1, 5.2, 4.7, 3.5, 9.7, 7.1],
    "O'sish":[0.3, 0.21, 0.4, 0.31, 0.5, 0.51, 0.62]
}
df = pd.DataFrame(dct)
df
```

```python
# numpy amalini DF ning har bir qiymati uchun qo'llash
np.abs(df)
```

```python
# DF ning barcha qiymatlarini butun qilib olish
np.round(df)
```

```python
# apply yordamida DF ning barcha qiymatlariga istalgan funksiyani qo'llashimiz mumkin
df.apply(np.sum)
```

```python
# apply yordamida qaysidur funksiyani qator yoki ustun bo'yicha bajarilishini ko'rsatsegam bo'ladi
df.apply(sum, axis=1)
```

```python
# apply ichiga o'zimiz funksiya yozsegam bo'ladi
f = lambda x: x.max() - x.min()
df.apply(f)
```

```python
# applymap DF ning har bir elementiga funksiyani alohida alohida bajaradi
df.applymap(lambda x: x*x) # har bir elementni kvadratini hisobliydi
```
