```python
import pandas as pd
```

Kolleryasiya - ikki ustun o'rtasidagi bog'liqlikni aniqlab beradi

```python
data = {
    'age':[10,20,30,40,50,60],
    'year':[2010,2000,1990,1980,1970,1960],
}

df = pd.DataFrame(data)
df
```

```python
# DF ning ikki ustunini bir-biriga bo'lgan bog'liqligini tekshirish
df['age'].corr(df['year'])
```

```python
# DF ning barcha ustunlarini umumiy bog'liqliklarini tekshirish
df.corr()
```

```python
# DF ning barcha ustunlarini faqat bitta ustuni bilan bog'liqligini tekshirish
df.corrwith(df['year'])
```
