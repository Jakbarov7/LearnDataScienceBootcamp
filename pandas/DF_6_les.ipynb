```python
import pandas as pd
```

Qator va ustunlarni tashlab yuborish : (`drop`) yordamida
- istalgan indeksni tashavorish
- bir qancha qatorlarni tashavorish
- ustunlarni tashavorish
- asl DF ga barcha o'zgarishlarni saqlab qolish

```python
data = {
	'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
	'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
	'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]
}
df = pd.DataFrame(data)
df
```

```python
# istalgan indeksni tashavorish
df2 = df.drop(3) # 3-indeksni tashavoramiz
```

```python
df2
```

```python
# bir qancha qatorlarni tashavorish
df.drop([0, 2, 4])
```

```python
# ustunlarni tashavorish
df.drop(['Aholi soni', 'Temp'], axis='columns')
```

```python
df
```

```python
# Asl DataFrame(DF) ga barcha o'zgarishlarni saqlab olish
df.drop('Aholi soni', axis=1, inplace=True)
```
