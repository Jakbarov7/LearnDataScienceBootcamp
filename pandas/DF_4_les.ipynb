```python
import numpy as np
import pandas as pd
```

Indeks metodlar:
- `.isin()` - N indeksi boshqa indeks ichida bor yo'qligini tekshiradi
- `.difference()` - boshqa indeksdan farqini aniqliydi
- `.union()` - ikkala indekslarni bitta qilib chiqaradi
- `.isin()` -
- `.delete()` - indekslardan birortasini o'chirish
- `.drop()` - qiymat nomi bilan o'chirish
- `.insert()` - DF ning istalgan joyiga yangi qiymat qo'shish
- `.is_monotonic` - qiymatlar bir xilda o'sib yoki kamayib borishini tekshiradi agarda shunday bo'lsa True aks holda False beradi
- `.is_unique` - biror element yana bir bor takrorlanganmi yo'qmi aniqlashtiradi
- `.unique()` - takrorlamasdan chiqarb beradi

DataFrame da indekslar takrorlanishi mumkin . Bu bilan bir xil indekslarda har xil qiymatlar bo'ladi degani .

```python
usernames = {"anvar":"anvarbek223", "olim":"olimtoy", "maryam":"mary334455","javohir":"javohir445"}
emails = {"anvar":"anvar@gmail.com", "olim":"olim@mail.ru", "maryam":"mary33@gmail.com","javohir":"javohir445@gmail.com"}

data = {
    'username':usernames,
    'email':emails
}
df = pd.DataFrame(data)
df
```

```python
data2 = {
    'username':{"olim":"olimjon21", "maryam":"maryam2020"},
    'email':{"olim":"olimjon21@mail.ru", "maryam":"maryam20@gmail.com"}
}
df2 = pd.DataFrame(data2)
df2
```

```python
df2 = pd.concat([df, df2])
df2
```

```python
# indekslarini ko'ramiz
df2.index
```

```python
# ustunlarini ko'ramiz
df2.columns
```

```python
# ikkitalik indekslardan birini olib ko'ramiz
df2.loc['olim']
```

```python
# yangi indeks obyektini yaratish
idxlar = pd.Index(['robiya', 'anvar', 'hamida'])
idxlar
```

```python
# isin - idxlardagilar DF indekslarida bor yo'qligini aniqliydi
idxlar.isin(df2.index)
```

```python
# difference - indekslar farqlarini aniqliydi
idxlar.difference(df2.index)
```

```python
df2.index.difference(idxlar)
```

```python
# df ning indekslarini vrbga saqlab olib ishlatamiz
df_idx = df2.index
df_idx
```

```python
# union - indekslarni birlashtrb chiqaradi
df_idx.union(idxlar)
```

```python
# delete - biror indeksni o'chirish
df2.index.delete(2) # 1-maryamni o'chiradi
```

```python
# drop - indeks nomi bilan o'chirish
df2.drop('maryam')
```

```python
# insert - istalgan joyga yangi qiymat qo'shish
df2.index.insert(3, 'robiya')
```

```python
# is_monotonic - bi xilda o'sib borishini aniqlash
df2.index.is_monotonic_increasing
```

```python
# is_monotonic_decrising - bir xilda kamayib borishini aniqlash
df2.index.is_monotonic_decreasing
```

```python
# unique - takrorlangan bo'sayam takrorlanmas qilib chiqaradi
df2.index.unique()
```
