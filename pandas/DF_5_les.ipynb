```python
import numpy as np
import pandas as pd
```

Qayta indekslash: (reindex) yordamida
- indekslarni tartiblash
- keraksiz indekslarni tashlab yuborish
- NaN larni DF ning yaqin va ohirgi qiymatlari bilan to'ldirvorish
- kerakli ustunlarni opqolib qoganlarini tashavorish
- tasodifiy qatorlarni olish
- istalgan ustun va qatorlarni olish , qoganlarini tashavorish


```python
# DF yaratamiz
data = {
	'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
	'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
	'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]
}
df = pd.DataFrame(data)
df
```

```python
# indekslarini qiymatlarini yangilariga o'zgartiramiz
df.index = [18, 17, 16, 15, 14, 13, 12, 11]
```

```python
df
```

```python
# reindex yordamida indeks qiymatlarini tartiblash
idx = [18, 17, 16, 15, 14, 13, 12, 11]
# birinchi - indekslarni tartiblimiz
idx.sort()
# reindex bilan qayta indekslemiz
df.reindex(idx)
```

```python
# reindex bilan - keraksiz indekslarni tashavorish
df.reindex([11, 13, 15, 17]) # kerakli indekslar yoziladi
```

```python
# reindexda indekslar soni DF indekslari bn teng bo'lishi shartmas
df.reindex(index=range(10, 22)) #10-19-20-21-22 idxlarda qiymat yo'q
```

```python
# yuqoridagi NaN qiymatlarini yaqin qiymatlar bilan to'ldirvorish
df.reindex(index=range(10, 22), method='nearest')
```

```python
# NaN larni DF ning eng ohirgi qiymati bn to'ldirvorish
df.reindex(index=range(10, 22), method='ffill')
```

```python
# reindex bilan - kerak ustunlar olib qoganini tashavorish
df.reindex(columns=['Temp', 'Yil'])
```

```python
# tasodifiy N ta qatorni olish
df.sample()
```

```python
df.sample(3)
```

```python
# kerakli ustun va qatorlarni olib qoganini tashavorish
df.loc[[18, 16, 14, 12], ['Aholi soni']]
```
