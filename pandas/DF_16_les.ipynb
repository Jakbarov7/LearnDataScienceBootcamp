```python
# DF chaqirib olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df.head()
```

```python
# DF ning biror ustunidagi elementlarini takrorlanmas qilib olish
df['model'].unique()
```

```python
# biror ustundagi qiymatlar nechtadan ekanligini aniqlash
df['model'].value_counts()
```

```python
# biror ustundan faqat bizga kerakli qiymatlarni alohida DF qilib ajratib olish
ents = [' S Class', ' GLC Class']
filtr = df['model'].isin(ents)
```

```python
df[filtr].sample(10)
```
