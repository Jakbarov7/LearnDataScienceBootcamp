```python
import numpy as np
import pandas as pd
```

**tartiblash(sort) va reytinglash(ranking)**
1. Seriesni indeks bo'yicha tartiblash
2. Seriesni qiymatlari bo'yicha tartiblash
3. DF uchun indeks bo'yicha tartiblash
4. DF da ustunlar nomi bo'yicha tartiblash (alfabet bo'yicha tahlanadi) `df.sort_index(axis=1)` bunda inplace ni parametr qilib berishimiz mumkin
5. ascending - o'sishga teskari , parametri yordamida DF ning indekslarini va ustunlarini teskari tartiblashimiz mumkin `df.sort_index(ascending=False)` yoki `df.sort_index(axis=1, ascending=False)`
6. DF ning biror ustuni bo'yicha qiymatlarini (o'sish) tartiblash | by parametri `df.sort_values(by='ustun_nomi')` buni ham ascending bilan kamayish tartibida joylashtirishimiz mumkin
7. bir vaqtni o'zida DF ning bir necha ustunlarini qiymatlari bo'yicha tartiblash `by=[ustunlar]` ascending ishlatsek bo'ladi

```python
obj = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
obj
```

```python
# indeks bo'yicha tartiblash
obj.sort_index()
```

```python
obj
```

```python
# qiymat bo'yicha tartiblash
obj.sort_values()
```

```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
# indeks bo'yicha tartiblash
df.sort_index()
```

```python
# ustunlar nomini tartiblash
df.sort_index(axis=1)
```

```python
df.head()
```

```python
df.sort_index(axis=1, inplace=True)
```

```python
df.head()
```

```python
# jadvalning indekslarini va ustunlarini teskari tartiblash
df.sort_index(ascending=False)
```

```python
# jadval ustunlarini teskari tartiblash
df.sort_index(axis=1, ascending=False)
```

```python
# jadvalning biror ustunining qiymatlarini tartiblash
df.sort_values(by='year')
```

```python
df.sort_values(by='price')
```

```python
# ascending bilan yuqoridagini teskari tartiblash
df.sort_values(by='year', ascending=False)
```

```python
df.sort_values(by='price', ascending=False)
```

```python
# bir vaqtning o'zida bir necha ustunni qiymatlarini tartiblash
df.sort_values(by=['price', 'year'], ascending=True)
```
