```python
import pandas as pd
```

Tartiblash(sorting)

```python
data = {
	'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
	'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
	'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]
}
df = pd.DataFrame(data, index=[10, 9, 8, 7, 6, 5, 4, 3])
df
```

```python
# DF ni indekslarini o'sish tartibida joylashtirish
df.sort_index()
```

```python
# DF ni ustunlarini alfabet bo'yicha tartiblash
df.sort_index(axis='columns')
```

```python
# qatorlarni teskari qilib tartiblash
df.sort_index(ascending = False)
```

```python
# ustunlarini teskari qilib tartiblash
df.sort_index(axis='columns', ascending = False)
```

```python
# DF ning bir ustunining qiymatlarini tartiblash
df.sort_values(by = 'Aholi soni')
```

```python
# yuqoridagini teskari qilib ham tartiblashimiz mumkin
df.sort_values(by = 'Aholi soni', ascending=False)
```

```python
# bir vaqtda bir nechta ustunlar qiymatlarini tartiblash
df.sort_values(by = ['Yil', 'Temp'])
```

```python
# yuqoridagilarni teskari tartiblashimiz mumkin
df.sort_values(by = ['Yil', 'Temp'], ascending=False)
```
